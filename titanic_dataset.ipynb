```python
# numpy for numerical operations and pandas for data manipulation.
import numpy as np
import pandas as pd
```

```python
# Filters and displays rows where 'Age' is greater than 30.
print(df[df['Age']>30])
```

```python
# sklearn Model selection,LogisticRegression and metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df["Embarked"] = df["Embarked"].fillna("S")
```

```python
# Fills missing 'Cabin' values with 'unknown'.
df["Cabin"] = df["Cabin"].fillna("unknown")
```

```python
# Converts 'Sex' (male/female) and 'Embarked' (S/C/Q) categorical features to numerical representations
# fills any remaining NaNs in the entire DataFrame with 0.
df["Sex"]=df["Sex"].map({"male":0,"female":1})
df["Embarked"]=df["Embarked"].map({"S":0,"C":1,"Q":2})
df.fillna(0,inplace=True)
```

```python
#  dropping non-predictive column
x=df.drop(["Survived", "Name", "Ticket", "PassengerId", "Cabin"],axis=1)
y=df["Survived"]
```

```python
# Splits the data into training
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Initializes a Logistic Regression model with a maximum iteration limit of 1000 and then trains it on the training data.
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
# trained Logistic Regression model to make predictions on the test set features.
y_pred=model.predict(x_test)
```

```python
# prints the accuracy score of the model's predictions
print("accuracy:",accuracy_score(y_test,y_pred))
```

```python
#  confusion matrix  to evaluate model performance, then visualizes it using a seaborn heatmap.
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt="d")
plt.title("confusion matrix")
plt.show()
```
